```python
import pandas as pd
import os
```

```python
file = os.path.join("InternationalGDP.csv")

# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df
```

```python
# setting column Country as index"
df.set_index("Country Name", inplace=True)
df
```

```python
# selecting desired columns"
df_yrs=df[["Country Code","Indicator Name", "1995","1996","1997","1998","1999","2000","2001","2002","2003","2004","2005","2006", "2007","2008", "2009","2010","2011","2012", "2013","2014","2015","2016","2017"]]
df_yrs
```

```python
df_yrs.count()
```

```python
df_clean = df_yrs.fillna(0)
df_clean
```

```python
#Saving Output
output_file = os.path.join("Clean_InterGDP.csv")

df_clean.to_csv(output_file, index=False, header=True)
```
